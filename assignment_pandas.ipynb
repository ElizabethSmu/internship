**8. Среди кого больше доля зарабатывающих много (>50K): среди женатых или холостых мужчин (признак *marital-status*)? Женатыми считаем тех, у кого *marital-status* начинается с *Married* (Married-civ-spouse, Married-spouse-absent или Married-AF-spouse), остальных считаем холостыми.**

```python
df_reach_man = df[(df['salary'] == '>50K') & (df['sex'] == 'Male')]

married_part = df_reach_man[df_reach_man['marital-status'].str.contains(r'Married', case=False)].shape[0]/df_reach_man.shape[0]
single_part = ~(df_reach_man[df_reach_man['marital-status'].str.contains(r'Married', case=False)]).shape[0]/df_reach_man.shape[0]

if married_part > single_part:
    print('Доля зарабатывающих много больше среди женатых мужчин')
else:
    print('Доля зарабатывающих много больше среди холостых мужчин')
```

**9. Какое максимальное число часов человек работает в неделю (признак *hours-per-week*)? Сколько людей работают такое количество часов и каков среди них процент зарабатывающих много?**

```python
h_max = df['hours-per-week'].max()
work_a_lot = df[df['hours-per-week'] == h_max]

print(f'Максимальное число рабочих часов в неделю: {h_max}')
print(f"Количество человек, работающих {h_max} часов в неделю: {len(work_a_lot)}")
print(f"Процент много зарабатывающих среди людей, работающих {h_max} часов в неделю: {len(work_a_lot[work_a_lot['salary'] == '>50K']) / len(work_a_lot[work_a_lot['salary'] == '<=50K'])*100:.2f} %")
```

**10. Посчитайте среднее время работы (*hours-per-week*) зарабатывающих мало и много (*salary*) для каждой страны (*native-country*).**

```python
df_avg_hours = df.groupby(['native-country', 'salary']).agg({'hours-per-week': 'mean'}).reset_index()
df_avg_hours_p = df_avg_hours.pivot(index='native-country', columns='salary', values='hours-per-week')
df_avg_hours_p.columns.name = None
df_avg_hours_p = df_avg_hours_p.reset_index()
df_avg_hours_p = df_avg_hours_p.rename(columns = {'<=50K': 'salary <=50K',
                                       '>50K': 'salary >50K'})
df_avg_hours_p
```

**11.Сгруппируйте людей по возрастным группам *young*, *adult*, *retiree*, где:**
* *young* соответствует 16-35 лет
* *adult* - 35-70 лет
* *retiree* - 70-100 лет

**Проставьте название соответсвтуещей группы для каждого человека в новой колонке AgeGroup**

```python
df['AgeGroup'] = df['age'].apply(lambda x: 'young' if 16 <= x < 35 else
                                           'adult' if 35 <= x < 70 else
                                           'retiree')
df.head()
```

**12-13. Определите количество зарабатывающих >50K в каждой из возрастных групп (колонка AgeGroup), а также выведите название возрастной группы, в которой чаще зарабатывают больше 50К (>50K)**

```python
df_a = df[df['salary'] == '>50K']
df_age_gr = df_a.groupby('AgeGroup').size().reset_index()
df_age_gr  = df_age_gr.rename(columns = {0:'count'}).sort_values(by='count', ascending=False)
df_age_gr
```

```python
print(f"Чаще всего зарабатывают больше 50К в возрастной группе {df_age_gr['AgeGroup'][0]}")
```

**14. Сгруппируйте людей по типу занятости (колонка occupation) и определите количество людей в каждой группе. После чего напишите функциюю фильтрации filter_func, которая будет возвращать только те группы, в которых средний возраст (колонка age) не больше 40 и в которых все работники отрабатывают более 5 часов в неделю (колонка hours-per-week)**

```python
df_occupation = df.groupby('occupation').agg({'fnlwgt':'count',
                                              'age':'mean',
                                              'hours-per-week':'min'}).reset_index()
df_occupation = df_occupation.rename(columns={'fnlwgt':'count',
                                              'age':'avg_age',
                                              'hours-per-week':'min_working_hours'})
df_occupation
```

```python
def filter_func(data, max_age = 40, min_hours = 5):
    return data[(data['avg_age'] <= max_age) & (data['min_working_hours'] >= min_hours)]['occupation'].tolist()
```

```python
occupation_list = filter_func(df_occupation)

print(*occupation_list)
```
